# TensorFlow / Keras Neural Network on MNIST

This notebook demonstrates:
- Loading and visualizing MNIST data
- Building a neural network with TensorFlow/Keras
- Training and evaluation
- Plotting loss and accuracy metrics
- Displaying predictions

```
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
import numpy as np

print("TensorFlow version:", tf.__version__)
```

## Load and Visualize Dataset

```
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

x_train = x_train / 255.0
x_test = x_test / 255.0

print("Training shape:", x_train.shape)
print("Testing shape:", x_test.shape)
```

```
fig, axes = plt.subplots(2, 5, figsize=(12, 5))

for i, ax in enumerate(axes.flat):
    ax.imshow(x_train[i], cmap='gray')
    ax.set_title(f"Label: {y_train[i]}")
    ax.axis('off')

plt.tight_layout()
plt.show()
```

## Build Model

```
model_tf = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dense(128, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

model_tf.summary()
```

```
model_tf.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

## Train Model

```
history = model_tf.fit(
    x_train,
    y_train,
    epochs=5,
    validation_split=0.1
)
```

## Plot Metrics

```
plt.figure(figsize=(8,5))
plt.plot(history.history['loss'], marker='o')
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("TensorFlow Training Loss")
plt.grid(True)
plt.show()
```

```
plt.figure(figsize=(8,5))
plt.plot(history.history['accuracy'], marker='o', label='Training Accuracy')
plt.plot(history.history['val_accuracy'], marker='o', label='Validation Accuracy')
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("TensorFlow Accuracy")
plt.legend()
plt.grid(True)
plt.show()
```

## Evaluate Model

```
test_loss, test_accuracy = model_tf.evaluate(x_test, y_test)

print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
```

## Display Predictions

```
predictions = model_tf.predict(x_test[:10])

fig, axes = plt.subplots(2, 5, figsize=(12, 5))

for i, ax in enumerate(axes.flat):
    pred = np.argmax(predictions[i])

    ax.imshow(x_test[i], cmap='gray')
    ax.set_title(f"True: {y_test[i]}\nPred: {pred}")
    ax.axis('off')

plt.tight_layout()
plt.show()
```
