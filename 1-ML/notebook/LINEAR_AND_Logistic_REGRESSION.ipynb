# Linear Regression From Scratch

## What is Linear Regression?

Linear Regression is one of the simplest and most important Machine Learning algorithms.

It tries to find the best linear relationship between:

- Input features $x$
- Target values $y$

The model predicts values using a straight line equation.



# Linear Regression Equation

$$
\hat{y} = wx + b
$$

Where:

| Symbol | Meaning |
|---|---|
| $\hat{y}$ | Predicted value |
| $w$ | Weight / Slope |
| $x$ | Input feature |
| $b$ | Bias / Intercept |



# Goal of Linear Regression

The goal is to find the optimal values of:

- $w$
- $b$

that minimize prediction error.



# Cost Function (Mean Squared Error)

We use Mean Squared Error (MSE):

$$
J(w,b) = \frac{1}{n}\sum_{i=1}^{n}(y_i - \hat{y}_i)^2
$$

Since:

$$
\hat{y}_i = wx_i + b
$$

the equation becomes:

$$
J(w,b) = \frac{1}{n}\sum_{i=1}^{n}(y_i - (wx_i + b))^2
$$



# Gradient Descent Update Rules

## Weight Update

$$
w := w - \alpha \frac{\partial J}{\partial w}
$$



## Bias Update

$$
b := b - \alpha \frac{\partial J}{\partial b}
$$



# Gradient Equations

## Derivative with respect to weight

$$
\frac{\partial J}{\partial w}
=
-\frac{2}{n}\sum x_i(y_i - \hat{y}_i)
$$



## Derivative with respect to bias

$$
\frac{\partial J}{\partial b}
=
-\frac{2}{n}\sum (y_i - \hat{y}_i)
$$



# Step-by-Step Implementation

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Training Data

```python
X = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([5, 7, 10, 11, 13], dtype=float)
```

## Initialize Parameters


```python
w = 0.0
b = 0.0
```

## Hyperparameters

```python
learning_rate = 0.01
epochs = 1000
```

## Number of samples

```python
n = len(X)
```

## Store loss history

```python
losses = []
```

## Gradient Descent Training

```python

for epoch in range(epochs):

    # Predictions
    y_pred = w * X + b

    # Compute Loss (MSE)
    loss = np.mean((y - y_pred) ** 2)

    # Store loss
    losses.append(loss)

    # Compute Gradients
    dw = (-2 / n) * np.sum(X * (y - y_pred))
    db = (-2 / n) * np.sum(y - y_pred)

    # Update Parameters
    w = w - learning_rate * dw
    b = b - learning_rate * db

    # Print progress
    if epoch % 100 == 0:
        print(f"Epoch {epoch}: Loss = {loss:.4f}")
```

## Final Parameters

```python
print("\nFinal Parameters")
print(f"Weight (w): {w:.4f}")
print(f"Bias (b): {b:.4f}")
```

## Predictions after training

```python
y_pred = w * X + b

# Plot data points
plt.figure(figsize=(8,5))

plt.scatter(X, y, label="Actual Data")

# Plot regression line
plt.plot(X, y_pred, label="Regression Line")

plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression From Scratch")

plt.legend()

plt.show()
```

# Logistic Regression From Scratch

## What is Logistic Regression?

Logistic Regression is a supervised Machine Learning algorithm used for:

- Binary Classification

It predicts probabilities between:

$$
0 \rightarrow 1
$$

Examples:

- Spam / Not Spam
- Fraud / Not Fraud
- Disease / No Disease
- Employee Leave / Stay



# Linear Equation

Logistic Regression first computes a linear equation:

$$
z = wx + b
$$

Where:

| Symbol | Meaning |
|---|---|
| $z$ | Linear output |
| $w$ | Weight |
| $x$ | Input feature |
| $b$ | Bias |

---

# Sigmoid Function

The linear output is passed through the Sigmoid function:

$$
\sigma(z) = \frac{1}{1 + e^{-z}}
$$

The sigmoid converts values into probabilities between:

$$
0 \leq \hat{y} \leq 1
$$



# Final Prediction Formula

$$
\hat{y} = \sigma(wx + b)
$$

Expanded:

$$
\hat{y} =
\frac{1}{1 + e^{-(wx+b)}}
$$



# Probability Interpretation

| Probability | Class |
|---|---|
| $\hat{y} \geq 0.5$ | Class 1 |
| $\hat{y} < 0.5$ | Class 0 |



# Cost Function (Binary Cross Entropy)

Logistic Regression uses Log Loss:

$$
J(w,b)
=
-\frac{1}{n}
\sum_{i=1}^{n}
\left[
y_i \log(\hat{y}_i)
+
(1-y_i)\log(1-\hat{y}_i)
\right]
$$



# Gradient Descent Update Rules

## Weight Update

$$
w := w - \alpha \frac{\partial J}{\partial w}
$$



## Bias Update

$$
b := b - \alpha \frac{\partial J}{\partial b}
$$



# Gradient Equations

## Weight Gradient

$$
\frac{\partial J}{\partial w}
=
\frac{1}{n}
\sum x_i(\hat{y}_i - y_i)
$$



## Bias Gradient

$$
\frac{\partial J}{\partial b}
=
\frac{1}{n}
\sum (\hat{y}_i - y_i)
$$


# Sigmoid Function Visualization

```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid Function

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Generate Input Values

z = np.linspace(-10, 10, 500)

# Compute Sigmoid Output
y = sigmoid(z)

# Plot
plt.figure(figsize=(10,6))

plt.plot(z, y)

# Reference line
plt.axhline(y=0.5, linestyle="--")

# Labels
plt.xlabel("z")
plt.ylabel("Sigmoid(z)")

plt.title("Sigmoid Function")

plt.grid(True)

plt.show()
```

# Step-by-Step Implementation

```python
import numpy as np
import matplotlib.pyplot as plt

# Training Data
X = np.array([1, 2, 3, 4, 5, 6], dtype=float)

y = np.array([0, 0, 0, 1, 1, 1], dtype=float)

# Initialize Parameters

w = 0.0
b = 0.0

# Hyperparameters
learning_rate = 0.1
epochs = 1000

# Number of samples
n = len(X)

# Store losses
losses = []


# Sigmoid Function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Training Loop
for epoch in range(epochs):

    # Linear equation
    z = w * X + b

    # Predictions
    y_pred = sigmoid(z)

    # Compute Log Loss
    loss = -np.mean(
        y * np.log(y_pred + 1e-9)
        +
        (1 - y) * np.log(1 - y_pred + 1e-9)
    )

    losses.append(loss)

    # Compute gradients
    dw = (1 / n) * np.sum(X * (y_pred - y))
    db = (1 / n) * np.sum(y_pred - y)

    # Update parameters
    w = w - learning_rate * dw
    b = b - learning_rate * db

    # Print progress
    if epoch % 100 == 0:
        print(f"Epoch {epoch}: Loss = {loss:.4f}")

# Final Parameters
print("\nFinal Parameters")
print(f"Weight (w): {w:.4f}")
print(f"Bias (b): {b:.4f}")
```

```python
# Test value
x_test = 4.5

# Linear equation
z = w * x_test + b

# Probability
probability = sigmoid(z)

# Predicted class
prediction = 1 if probability >= 0.5 else 0

print(f"Probability: {probability:.4f}")
print(f"Prediction: {prediction}")
```

```python
plt.figure(figsize=(8,5))

plt.plot(losses)

plt.xlabel("Epoch")
plt.ylabel("Loss")

plt.title("Loss Reduction During Training")

plt.show()
```
